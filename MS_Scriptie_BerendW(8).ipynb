#### Some functions that help with extracting words from text

````python
import re

def extract_string_between_tags(string):
    pattern = r'<pad> (.*?)</s>'
    match = re.search(pattern, string)
    if match:
        extracted_string = match.group(1).strip()
        return extracted_string
    else:
        return None

def remove_x_followed_by_integers(string):
    pattern = r'\x\d{2}'
    result = re.sub(pattern, '', string)
    return result

def contains_x_followed_by_integers(string):
    if '\\x' in string:
        index = string.index('\\x')
        if len(string) >= index + 4:
            if string[index + 2].isdigit() and string[index + 3].isdigit():
                return True
    return False

def check_quotes(string):
    count = 0
    for char in string:
        if char == "'":
            count += 1
            if count == 2 or count ==4:
                return True
    return False

import re

def extract_text(string):
    pattern = r"(?:(?<=falls under the ')|(?<=falls under the category of '))([^']*)"
    match = re.search(pattern, string)
    if match:
        extracted_text = match.group(1)
    else:
        extracted_text = ""

    return extracted_text

import re

def extract_text2(string):
    pattern = r"(?<=falls under the category of )\w+"
    match = re.search(pattern, string)
    extracted_text = ""

    if match:
        extracted_text = match.group(0)

    return extracted_text

def c(sentence, target_phrase):
    words = sentence.split()
    try:
        index = words.index(target_phrase)
        if index + 1 < len(words):
            next_word = words[index + 1]
            return next_word
    except ValueError:
        pass
    return None


````

````python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from typing import List
import re
model = AutoModelForSeq2SeqLM.from_pretrained("google/flan-t5-xxl",low_cpu_mem_usage=True)
tokenizer = AutoTokenizer.from_pretrained("google/flan-t5-xxl",low_cpu_mem_usage=True)

class FlanT5XXL:
  """
  This class represents all LLM model called Flan-T5-XXl created by google:
  https://huggingface.co/google/flan-t5-xxl
  Answers of the model come out the following way:
    <pad> Household Kitchen Merchandise</s>

  """

  def __init__(self, json_title:str,json_text:str,json_specs:str) -> None:
    """
    This method is the initialising method, that is always called when the class is created.
    param json_title: the title a itme from the json file contains
    param json_text: the text a item from the json file contains
    param json_specs: the specs a item from the json file contains
    """
    self.json_title = json_title
    self.json_text = json_text
    self.json_specs = json_specs


  def loop_1(self,list_names:List, list_with_root_nodes:List) -> str:

    """
    This method gives the root node that the product belongs to, returns a string containing the hierachy value of the next node
    param list_names: a list with the names of root nodes
    param list_with_root_nodes: a list with the root nodes and the hierachy values
    param return: returns a string containing a hierachy value
    """

    input_text = f"Under which product taxonomy category does: '{self.json_title}' Given the follwing categories: {list_names}?"
    input_ids = tokenizer(input_text, return_tensors="pt").input_ids
    outputs = model.generate(input_ids,  max_new_tokens=400, temperature=0.7, do_sample=True)
    new = tokenizer.decode(outputs[0]).strip('<pad> ').strip('</s')

    print(self.json_title)
    for x in list_with_root_nodes:
      if new in x[0]:
        next_node = x[1]
    return next_node

  def long_loop(self,new_node:str) -> str:

    """
    This method iterates multiple times, till the the final leaf node is found, returns the final category
    param new_node: the node found by def loop_1
    param return: returns a string containing the final category.
    """

    fail_safe_counter =0
    next_node = new_node
    while True:
      lis =[x.encode("ascii", "ignore").decode() for x in subClassesOfSubject(next_node).keys()]
      numbers = [(x[0].encode("ascii", "ignore").decode(),x[1]) for x in subClassesOfSubject(next_node).items()]
      print(lis)
      print(numbers)
      input_text = f"Under which product taxonomy category does: '{self.json_title}' fall? Choose one of follwing categories: {lis}."
      input_ids = tokenizer(input_text,return_tensors="pt").input_ids
      outputs = model.generate(input_ids,  max_new_tokens=400, temperature=0.7, do_sample=True)
      print(tokenizer.decode(outputs[0]))

      if '\'' in tokenizer.decode(outputs[0]):
        matches = re.findall("'([^']*)'", tokenizer.decode(outputs[0]))
      else:
        matches = extract_string_between_tags(tokenizer.decode(outputs[0]))
      print(matches)

      if '<unk>' in matches:
        matches.replace('<unk>','\\')


      if len(numbers)<=1:
        for x in numbers:
          if type(matches) == type(list):
            if matches[0] in x:
              next_node= x[1]
              fail_safe_counter += 1
          else:
            if matches in x:
              next_node =x[1]
              fail_safe_counter += 1
      else:
        # numbers = numbers[0]
        for x in numbers:
          if matches in x:
            next_node= x[1]
            fail_safe_counter += 1
      print(next_node)

      #to check if they program chooses something that does not exist, it still keeps going
      # if type(matches) == type(list):
      #   if matches[0] not in lis:
      #     next_node =numbers[0][1]
      # if matches not in lis:
      #      next_node =numbers[0][1]

      if next_node == 'final_node':
        print('Final Node')
        if type(matches) == type(list):
          print(f'And the category is: {matches[0]}')
          return matches[0]
        else:
          print('And the category is:')
          return matches

      if next_node == '':
        print('Final Node')
        if type(matches) == type(list):
          print(f'And the category is: {matches[0]}')
          return matches[0]
        else:
          print('And the category is:')
          return matches

      if type(next_node) == type(list):
        if new_node[0][0] == '2':
          print('And the final category is:')
          return matches
        if next_node[0] == '':
          return matches

      if fail_safe_counter >8:
        return matches

  def attributes_extraction_via_text(self) -> str:
    """
    A method that returns the attributes from the json_text
    param return: json text
    """


    input_text = f"What attributes does: '{self.json_title}' have? Also give the attributes given in the text: {self.json_text}."
    # input_text = f"What attributes does: '{self.json_title}' have?

    input_ids = tokenizer(input_text,return_ensors="pt").input_ids
    outputs = model.generate(input_ids,  max_new_tokens=1000, temperature=0.7, do_sample=True)
    attributes = extract_string_between_tags(tokenizer.decode(outputs[0])).replace(' is ',' = ').split('-')
    return attributes
    # lis =[x for x in subClassesOfSubject(next_node).keys()]
    # numbers = [x for x in subClassesOfSubject(next_node).items()]
````
